Zadanie 1

Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:


*   wypełnianie deklaracji PIT, - wymaga inteligencji od człowieka

*   streszczanie tekstu, - Nie wymaga inteligencji od człowieka, ponieważ nie jest to trudne zadanie, nie wymaga wiedzy na żaden temat, zadanie wymaga tylko cztania i skracania tego co się przeczytało

*   tłumaczenie tekstu - wymaga inteligencji od człowieka, ponieważ po pierwsze trzeba znać dobrze dwa języki i trzeba wiedzieć jakie tłumaczenie najbardziej pasuje do danego słowa

*   klasyfikacja tekstu do kategorii tematycznych, - nie wymaga inteligencji od człowieka

*   odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim), - Nie wymaga inteligencji od człowieka jęsli są to proste pytania. np Ile godzin dziś spałaś? odpowiedz na to pytanie nie wymaga inteligencji


*   układanie rozkładu jazdy transportu miejskiego, wymaga inteligencji od człowieka



*   programowanie (pisanie programów komputerowych), - wymaga inteligencji od człowieka

*   „programowanie” kanałów telewizyjnych - wymaga inteligencji od człowieka

*  testowanie oprogramowania, - wymaga inteligencji od człowieka
*   komponowanie muzyki - Nie wymaga inteligencji od człowieka


*   rozwiązywanie układów równań, - wymaga inteligencji od człowieka


*   symboliczne obliczanie pochodnych funkcji - wymaga inteligencji od człowieka ponieważ sa to zadania które wymagaja analizy i logicznego myslenia w czym potrzebna jest inteligencja


*   symboliczne całkowanie funkcji - wymaga inteligencji od człowieka

*   kierowanie samochodem. - wymaga inteligencji od człowieka, ponieważ na drodze mogą zdarzyć się nieprzewidziane sytuacje, i trzeba wiedziec jak się zachować i analizować sytuacje na bierząco np. w ruchliwym mieście

Zadanie 3


Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):



*   rozmowa towarzyska, - Może być skutecznie imitowana przez sztuczne systemy, jest to dosyć proste zachowanie

*   dyskusja polityczna - Może być skutecznie imitowana

*   dyskusja naukowa - Nie może być skutecznie imitowana, poniewaz sztuczna inteligencja nie ma dostępu do wszystkich naukowych artykułów, a jak juz ma to jest to ograniczona ilość

*   odpowiadanie na pytania klientów w telefonicznej infolinii, - Nie może, ponieważ nawet jęsli sztuczna inteligencja znałaby odpowiedzi na niektóre pytania i miałaby troche informacji to nigdy nie można przewidzieć o co zapyta się drugi człowiek, może zadać pytanie  na  taki temat, na który sztuczna inteligencja nie ma danych

*   odpowiadanie na pytania klientów w internetowej infolinii - to samo co na górze
