Zadanie 1

Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:


*   wypełnianie deklaracji PIT, - - nie wymaga, ponieważ polega jedynie na podstawieniu danych wartości w odpowiednie komórki w tabelce według pewnych już ustalonych, podanych reguł.

*   streszczanie tekstu, - Nie wymaga inteligencji od człowieka, ponieważ nie jest to trudne zadanie, nie wymaga wiedzy na żaden temat, zadanie wymaga tylko cztania i skracania tego co się przeczytało

*   tłumaczenie tekstu - wymaga inteligencji od człowieka, ponieważ trzeba znać dobrze dwa języki i trzeba wiedzieć jakie tłumaczenie najbardziej pasuje do danego słowa

*   klasyfikacja tekstu do kategorii tematycznych, - nie wymaga, gdyż dopasowanie słów kluczowych odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim) - zależy od definicji prostego pytania, jeśli rozumiane jest przez to typowe, częste pytanie w danej sytuacji to nie jest potrzebna, natomiast nawet proste składniowo pytania mogą mieć skomplikowaną, nieprzewidywalną odpowiedź

*   odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim), - Nie wymaga inteligencji od człowieka jęsli są to proste pytania. np Ile godzin dziś spałaś? odpowiedz na to pytanie nie wymaga inteligencji


*   układanie rozkładu jazdy transportu miejskiego,  zależy - Można to zrobić za pomocą algorytmów optymalizacyjnych, ale bez inteligentnych metod może to działać wolno i słabo reagować na zmiany, takie jak nietypowe sytuacje, potrzeby pasażerów czy warunki w danym miejscu. Gdy jest wiele zmiennych, takie tradycyjne metody mogą być trudne do zastosowania lub nawet niemożliwe



*   programowanie (pisanie programów komputerowych), - wymaga, gdyż przed napisaniem programu potrzebny jest pomysł, stworzenie jego zamysłu, umiejętność twórczego rozwiązywania problemów

*   „programowanie” kanałów telewizyjnych - nie wymaga, przypisywanie do przycisków pilota odpowiednich programów odbywa się czysto schematowo

*  testowanie oprogramowania, - wymaga, potrzebne jest nieszablonowe myślenie, aby sprawdzić i dojść do niespotykanych sytuacji, danych, tworzenie scenariuszy testów
*   komponowanie muzyki - wymaga, ponieważ jest to proces oparty na kreatywności i twórczym myśleniu, często związany z emocjami, wrażliwością estetyczną oraz potrzebą tworzenia czegoś oryginalnego, wykraczającego poza utarte schematy i powszechnie przyjęte zasady


*   rozwiązywanie układów równań, - Zależy przy prostych zadaniach wystarczy działanie według ustalonych schematów, ale przy bardziej złożonych problemach brakuje gotowych rozwiązań, więc konieczna jest kreatywność


*   symboliczne obliczanie pochodnych funkcji -- nie wymaga potrzebne jest jedynie podstawienie odpowiednich wartości do wzoru według określonych reguł.


*   symboliczne całkowanie funkcji - Wymaga, ponieważ w niektórych sytuacjach nie ma gotowego wzoru, więc całkowanie staje się twórczym procesem prowadzącym do znalezienia odpowiedniego rozwiązania

*   kierowanie samochodem. - wymaga inteligencji od człowieka, ponieważ na drodze mogą zdarzyć się nieprzewidziane sytuacje, i trzeba wiedziec jak się zachować i analizować sytuacje na bierząco np. w ruchliwym mieście
